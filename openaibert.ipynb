```python
import torch
from transformers import BertForMaskedLM, BertTokenizer
```

```python
#inhitialize tokenizer and model from pretyrained BERT model
model_name = 'bert-base-uncased'  
model = BertForMaskedLM.from_pretrained(model_name)
tokenizer = BertTokenizer.from_pretrained(model_name)
```

```python

sequence = "Information technology strategy and policy subject teaches us the various uses of AI and their uses"
#This line tokenizes the input sequence using the tokenizer specific to each model (GPT2Tokenizer for GPT-2 and BertTokenizer for BERT). It returns the encoded tokens as input tensors for the respective models.
inputs = tokenizer.encode(sequence, return_tensors='pt')
outputs = model.generate(inputs, max_length=200, do_sample=True, num_return_sequences=1,temperature=1, pad_token_id=tokenizer.eos_token_id)

generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)


print(generated_text)
```
